```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

_ = sns.set_style('whitegrid')
```

```python
def bm25_weighted_tf(tf, k, b, d, d_avg):
    """bm25."""
    denominator = k * ( 1 - b + b * (d / d_avg) ) + tf
    return tf / denominator
```

# Varying k

```python
d = 60
d_avg = 50
b = 0

k_min, k_max, k_step = 0, 2., .3
tf_min, tf_max, tf_step = 1, 100, 5
results = []
for k in np.arange(k_min, k_max, k_step):
    results.append( (k, [bm25_weighted_tf(tf=x, k=k, b=b, d=d, d_avg=d_avg) for x in np.arange(tf_min, tf_max, tf_step)]) )

f, ax = plt.subplots(1, 1, figsize=(7, 5))
for k, ys in results:
    _ = sns.lineplot(x=np.arange(len(ys)), y=ys, ax=ax, label=k)
```

# Varying b

```python
d = 60
d_avg = 50
k = 1.

b_min, b_max, b_step = 0, 1., .5
tf_min, tf_max, tf_step = 1, 100, 5
results = []
for b in np.arange(b_min, b_max, b_step):
    results.append( (b, [bm25_weighted_tf(tf=x, k=k, b=b, d=d, d_avg=d_avg) for x in np.arange(tf_min, tf_max, tf_step)]) )

f, ax = plt.subplots(1, 1, figsize=(7, 5))
for b, ys in results:
    _ = sns.lineplot(x=np.arange(len(ys)), y=ys, ax=ax, label=b)
```
